Question 1 : What is Data Loading in Power BI and why is it considered the first step of analysis?




Data Loading means bringing data from Excel, CSV, database, or other sources into Power BI. It is the first step because we need data before we can analyze it or make charts and reports. Without loading data, Power BI cannot do any analysis.

Question 2 : Explain the difference between “Load” and “Transform Data” in Power BI.




**Load**: It directly imports the data into Power BI.
**Transform Data**: It opens Power Query where we can clean or change the data before loading it.

For example, if there are blank rows or wrong column names, we can use Transform Data to fix them before loading.

Question 3 : What is a Fact Table and a Dimension Table? Give examples from the dataset.




**Fact Table**: It contains numerical data like sales, quantity, or profit. It is the main table used for analysis.

Example: Sales table with Sales Amount and Quantity.

**Dimension Table**: It contains details about the data such as customer names, product names, or dates.
Example: Product table, Customer table, and Date table.


Fact tables store numbers, while dimension tables store descriptions.

Question 4 : Why is Star Schema preferred over Snowflake Schema in Power BI?

Star Schema is preferred because it is simple and easy to understand. It has one fact table connected directly to dimension tables. It also works faster in Power BI and makes reports easier to create.

Snowflake Schema has more tables and relationships, so it can be more complicated and sometimes slower.

Therefore, Star Schema is mostly used in Power BI because it is easier and gives better performance.

Question 5 : Identify and remove duplicate records based on Date, Country, and State.




We do this inside the Power Query Editor before loading the data.

1. Open Power Query Editor (click Transform Data).

2. Hold down the Ctrl key and click the headers for the Date, Country, and State columns to select all three at once.

3. Right-click on one of the selected headers.

4. Click Remove Duplicates.

Question 6 : Identify and replace null values in Vaccination_Status.


We can also do this easily in Power Query Editor.

1. Right-click on the header of the Vaccination_Status column.

2. Select Replace Values...

3. In the box that says Value To Find, type null (leave it blank if it's literally just empty cells).

4. In the box that says Replace With, type Unknown.

5. Click OK.
